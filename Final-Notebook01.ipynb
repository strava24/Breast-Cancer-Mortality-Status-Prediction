## Missing Value Imputation along with Outlier Detection

Before dealing with the missing values, it is better to deal with the outliers cause outliers can effect statistical imputations

Reference:

Seminar 01 (week 02): page 23

```python
# Visualizing data to find outliers in Age
age_fig = px.histogram(data, x='Age')
age_fig.show()
```

There are age values like -50, 180 and 500. These are extreme outliers which can never be possible practically, So removing those outliers

Reference:

Seminar 01 (week 02): page 31

```python
# Creating a mask for age outliers (negative or > 100) that doesn't include nulls
# Getting the outliers that are not null
outlier_mask = ((data['Age'] <= 0) | (data['Age'] > 100)) & pd.notna(data['Age'])

# Counting the actual outliers (excluding nulls)
actual_outliers = outlier_mask.sum()
print(f"Actual outlier ages (excluding nulls): {actual_outliers}")

# Applying the filter
data = data[~outlier_mask]

data.info()
```

Visualizing age again to make sure that the outliers have been removed

Reference:

Seminar 01 (week 02): page 23

```python
# Visualizing data to make sure that the outliers have been removed properlly
age_fig = px.histogram(data, x='Age')
age_fig.show()
```

In this histogram 89 might look like an outlier it is indeed a statistical outlier, but in medical setting being diagnosed with breast cancer at the age of 89 might be possible.

BCRF. (2025). Breast Cancer in the Elderly | BCRF. Breast Cancer Research Foundation. Available from https://www.bcrf.org/about-breast-cancer/breast-cancer-elderly/.

Imputing the missing values of Age use median because, median is less sensitive to any remaining outliers cause it is the 50th percentile

```python
# Now imputing missing values in the age
age_median = data['Age'].median()

# Impute missing values
data['Age'] = data['Age'].fillna(age_median)
data.info()
```

## Data Scaling

Have to scale the categorical feature variables with non-binary encoding (T_Stage, N_Stage, 6th_Stage, Differentiated) and ordinal variable (Grade)

Reason: Non-scaled variables can result in biased decision makings in KNN (Distance-based, needs scaled features to avoid bias), LR and NB (Assumes normal distribution, scale to meet that assumption)

StandardScaler to scale these variables because,
- Helps distance-based models (like KNN) treat all features fairly
- Helps Logistic Regression converge faster and make better predictions
In summary it is the best option for distance based models to have a non-bias prediction

Binary encoded categorical features don't require scaling cause they have no impact on Distance Metrics, and the distribution is not very huge comparing to the other features after scaling

Reference:

Seminar 01 (week 02): page 35

```python
from sklearn.preprocessing import StandardScaler

# Create a single scaler instance
scaler = StandardScaler()

# Define the categorical features to scale
categorical_features = ['T_Stage', 'N_Stage', '6th_Stage', 'Differentiated', 'Grade']

# Apply scaling to all these features at once
data[categorical_features] = scaler.fit_transform(data[categorical_features])
```

Doing Standard Scaling on the remaining numeric/continuous features too for the above mentioned same reasons

Reference:

Seminar 01 (week 02): page 35

```python
# Define remaining numeric features
numeric_features = ['Age', 'Tumor_Size', 'Regional_Node_Examined', 'Reginol_Node_Positive']

# Create a single scaler for numeric features
numeric_scaler = StandardScaler()

# Apply scaling to all numeric features at once
data[numeric_features] = numeric_scaler.fit_transform(data[numeric_features])
```
